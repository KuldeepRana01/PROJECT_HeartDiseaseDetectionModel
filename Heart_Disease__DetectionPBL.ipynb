# **HEART DISEASE PREDICTION MODEL:**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #metrics form evaluting my model performance
```

```python
df = pd.read_csv('/content/heart_disease_data.csv') #lOAD DATASET
```

```python
df['target'].value_counts() #1-defective heart/0-healthy heart
```

Represent by number
* 1 : Defective Heart
* 0 : Healthy Heart


```python
X = df.drop(columns='target',axis=1)
```

```python
X
```

```python
Y = df['target']
```

```python
Y
```

**SPLITTING DATASET INTO TRAIN AND TEST**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X_test.shape, X_train.shape)
```

#  MODEL TRAINING

```python
model = LogisticRegression()
```

```python
#Train the model with Training data
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
Y_train
```

```python
score = accuracy_score(Y_test,y_pred)
```

```python
print("The accuracy score is ",score)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression # Assuming the model is already trained

def predict_heart_health():
  """Predicts heart health based on user input."""


  age = float(input("Enter your age: ")) # Get input data from the user
  cpk = float(input("Enter your CPK levels: "))
  serum_cholestrol = float(input("Enter your serum cholestrol levels: "))
  max_heart_rate = float(input("Enter your maximum heart rate achieved: "))
  oldpeak = float(input("Enter your oldpeak value: "))
  slope = float(input("Enter your slope value: "))
  ca = float(input("Enter your number of major vessels (0-3): "))
  thal = float(input("Enter your thal value (1 = normal; 2 = fixed defect; 3 = reversable defect): "))
  sex = float(input("Enter your sex (1 = male; 0 = female): "))
  fbs = float(input("Enter your fasting blood sugar (1 = >120 mg/dl; 0 = <120 mg/dl): "))
  resting_bp = float(input("Enter your resting blood pressure: "))
  exang = float(input("Enter your exercise induced angina (1 = yes; 0 = no): "))
  restecg = float(input("Enter your resting electrocardiographic results (0 = normal; 1 = having ST-T wave abnormality; 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria): "))



  input_data = (age,cpk,serum_cholestrol,max_heart_rate,oldpeak,slope,ca,thal,sex,fbs,resting_bp,exang,restecg)
  arr = np.asarray(input_data)  # Change the input_Data into array


  arr_reshape = arr.reshape(1, -1) # Reshape the array


  prediction = model.predict(arr_reshape) # Make prediction


  if prediction == 0: # Print the prediction
    print('You are healthy')
  else:
    print("Consult your doctor")

# Call the function to start the prediction process
predict_heart_health()
```

# **Now Prediction by Decision Tree..**


```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier  # Import DecisionTreeClassifier

def predict_heart_health():
  """Predicts heart health based on user input."""

  # Get input data from the user
  age = float(input("Enter your age: "))
  cpk = float(input("Enter your CPK levels: "))
  serum_cholestrol = float(input("Enter your serum cholestrol levels: "))
  max_heart_rate = float(input("Enter your maximum heart rate achieved: "))
  oldpeak = float(input("Enter your oldpeak value: "))
  slope = float(input("Enter your slope value: "))
  ca = float(input("Enter your number of major vessels (0-3): "))
  thal = float(input("Enter your thal value (1 = normal; 2 = fixed defect; 3 = reversable defect): "))
  sex = float(input("Enter your sex (1 = male; 0 = female): "))
  fbs = float(input("Enter your fasting blood sugar (1 = >120 mg/dl; 0 = <120 mg/dl): "))
  resting_bp = float(input("Enter your resting blood pressure: "))
  exang = float(input("Enter your exercise induced angina (1 = yes; 0 = no): "))
  restecg = float(input("Enter your resting electrocardiographic results (0 = normal; 1 = having ST-T wave abnormality; 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria): "))

  # Change the input_Data into array
  input_data = (age, cpk, serum_cholestrol, max_heart_rate, oldpeak, slope, ca, thal, sex, fbs, resting_bp, exang, restecg)
  arr = np.asarray(input_data)

  # Reshape the array
  arr_reshape = arr.reshape(1, -1)

  # Make prediction using Decision Tree
  prediction = model.predict(arr_reshape)  # Assuming 'model' is a trained DecisionTreeClassifier

  # Print the prediction
  if prediction == 0:
    print('You are healthy')
  else:
    print("Consult your doctor")

# Call the function to start the prediction process
predict_heart_health()
```
